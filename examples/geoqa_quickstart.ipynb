# GeoQA Quickstart

This notebook shows one small end-to-end GeoQA path:

1. Load a local sample dataset
2. Run one geometry validator
3. Write a JSON report
4. Read the report back for inspection


```python
from pathlib import Path
import json

import geopandas as gpd

from geoqa.reports.report_generator import generate_report
from geoqa.validations.geometry import duplicate_vertex
```

```python
dataset_path = Path("data/public_samples/edge_cases/duplicate_vertex_line.geojson")
report_base = Path("data/integration_results/quickstart_duplicate_vertex_report")

dataset_path, report_base
```

```python
layer = gpd.read_file(dataset_path)
layer
```

```python
issues = duplicate_vertex(layer)
len(issues)
```

```python
[issue.to_dict() for issue in issues]
```

```python
generate_report(
    issues,
    output_format="json",
    file_path=str(report_base),
)

report_base.with_suffix(".json")
```

```python
with report_base.with_suffix(".json").open("r", encoding="utf-8") as handle:
    report_payload = json.load(handle)

report_payload
```
